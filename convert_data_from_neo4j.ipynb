### Data Augmentation

### Ingredients

#### Getting all unique values of Ingredients

```python
df_recipes.head(1)
```

```python
display(f'Unique ingredients: {len(df_recipes['ingredients'].unique())}')
```

```python
df_ingredients = df_recipes['ingredients'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).to_frame('ingredients')
```

```python
df_ingredients.head()
```

#### Remove Numbers

```python
import re
def remove_numbers(text):
    return re.sub(r'\d+', '', text)
```

```python
df_ingredients['ingredients'] = df_ingredients['ingredients'].apply(remove_numbers)
```

```python
df_ingredients['ingredients']
```

#### Remove Unit Measurement

```python
units = ['Xícara', 'Litro', 'Copo Americano', 'Pitada', 'Colher de Chá', 'Colher de Sopa']
def remove_units(text, units):
    for unit in units:
        text = text.replace(unit, '')
    return text
```

```python
df_ingredients['ingredients'] = df_ingredients['ingredients'].apply(lambda x: remove_units(x, units))
```

```python
df_ingredients['ingredients']
```

#### Remove Special Characters

```python
special_characters = '[!"#$%°&\'()*+,-./:;<=>?@[\\]^_`{|}~]'
def remove_special_characters(df, column_name, characters_to_remove):
    df[column_name] = df[column_name].str.replace(f'[{re.escape(characters_to_remove)}]', ' ', regex=True)
    return df
```

```python
df_ingredients = remove_special_characters(df_ingredients, 'ingredients', special_characters)
```

```python
df_ingredients['ingredients'] = df_ingredients['ingredients'].str.strip()
```

```python
df_ingredients
```

#### Remove Articles, Prepositions

```python
prepositions = ['De ', 'Do ', 'Da ']
def remove_prepositions(text):
    for preposition in prepositions:
        text = text.replace(preposition, '')
    return text
```

```python
df_ingredients['ingredients'] = df_ingredients['ingredients'].apply(remove_prepositions)
```

```python
df_ingredients['ingredients']
```

#### Adding more ingredients to recipe

```python
import random
ingredients = df_ingredients['ingredients'].unique().tolist()

def get_4_ingredients(x):
    ing = random.choices(ingredients, k=4)
    ing = str(ing).replace('[', '').replace(']', '').replace("'", '').replace(',', '')
    return ing
```

```python
df_recipes['ingredients'] = df_recipes['ingredients'].apply(get_4_ingredients).astype(str)
```
