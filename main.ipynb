### Importando a base de dados

```python
import pandas as pd
import json
```

```python
with open("dataset/records.json") as f:
    data = f.read()
```

```python
data = data.replace("ï»¿", "")
```

```python
json_data = json.loads(data)
```

```python
df = pd.json_normalize(json_data)
```

```python
df.head()
```

```python
df.info()
```

### Separando o que é Receita, Usuário e Tag

#### Tratando as labels...

```python
df["n.labels"] = df["n.labels"].astype("string")
```

```python
df['n.labels'].value_counts()
```

#### Dataframe de Receitas

```python
df_recipes = df[df["n.labels"] == "['recipe']"]
df_recipes
```

#### Dataframe de Usuários

```python
df_users = df[df["n.labels"] == "['user']"]
df_users
```

```python
df_users['n.identity'].unique()
```

#### Dataframe de Tags

```python
df_tags = df[df["n.labels"] == "['tag']"]
df_tags
```

```python
df_tags['n.properties.label'].unique()
```
